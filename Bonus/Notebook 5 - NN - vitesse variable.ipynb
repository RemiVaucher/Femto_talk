# Schéma avec réseau de neurones - vitesse variable : 

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
 model = tf.keras.models.load_model("model2_1")
```

Dans le schéma défini nous récupérons de notre réseaux de neurones les  $\alpha^n _{j+1/2}$ ($\alpha \beta$) à l'instar du cas avec une vitesse constante.

Nous étudions différentes stratégies pour prendre en compte une vitesse variable.
- La première est la plus simple. Elle consiste à prendre en compte une vitesse constante sur l'ensemble de la masse mais qui varie au cours du temps .

- La seconde quand à elle va prendre une vitesse variable sur l'espace (ou peut se le représenter par plusieurs courant marins sur une roche dans l'eau par exemple, doit provoquer une répartition de la quantité de matère). On établit ainsi qu'au cours du temps les vitesses sur chaque maille ne changent pas.  

- La troisième consiste à prendre en compte une vitesse qui évolue en fonction de la quantité de matière sur chaque maille.



# Vitesse variable

### Premier cas :  Vitesse qui varie seulement au cours du temps

Pour ce faire, on définit par exemple la fonction suivante :

```python
def u(t,x): 
    ''' 
    vitesse variable par rapport au temps
    t n'excédant pas 1
    '''
    if t < 1/2 : return 0.1
    if t > 1/2 : return 0.8
```

Une grande difficulté rencontrée dans cette section, a été le passage de la masse du bord droit au bord gauche.


```python
N_t = 80
N_x = 40
dt= 1/(N_t-1)
dx=1/(N_x-1) 
x=np.linspace(0,1,N_x)

# Je fais juste accelerer la vitesse au cours du temps : 
# u vitesse 
t= dt
#beta = u(t,x)* dt/dx

#initialisation avec la fonction indicatrice
I_inf=0.2
I_sup=0.4
a0 = np.zeros(N_x)
a0[x>I_inf]=1 
a0[x>I_sup]=0

#Pour commencer
predict_1 =model.predict(np.append(a0[0:3],u(t,x)*dt/dx).reshape(1,4))
alpha=np.zeros((N_t,N_x))
alpha[0,:]=a0 
```

```python
# Tant que la condition CFL est validée on observe le déplacement de matière suivant :  
```

A un même temps j'ai la même vitesse sur l'ensemble de la masse considérée.

```python
n=1
cfl = max(a0)*dt/dx
print(cfl)
pred = predict_1[0,:] #les trois qté de matière pour premiere maille
stock = 0 #pred[0] Je me le permet ici car on a 0 quantité de matière dans la premiere maille.
pred = pred[1]
while cfl<1 and n<N_t-1: # La condition cfl doit être controlée pour continuer les itérations
    if n%10==1:
        print(n) #Pour suivi du temps d'exécution
    for i in range(0,N_x-2): 
        alpha[n,i+1]=alpha[n-1,i] - pred + stock      
        #Je prédis la quantité de matière sortante de la maille suivante
        stock = pred
        pred = model.predict(np.append(alpha[n-1,i:][0:3],u(t,x)*dt/dx).reshape(1,4))[0,:][1] #sur n-1 pour beta aussi ?
    cfl = max(alpha[n,:])*dt/dx
    #Conditions bords :Si on veut les conditions périodiques : alpha[n,0]= alpha[n,N_x-1] =0
    n=n+1
    alpha[n,0]=alpha[n-1,N_x-2] # Pour forcer le passage du bord 
    alpha[n,1]=alpha[n-1,N_x-1] 
    t+=dt
X = np.linspace(0,1,N_x)
```

```python
l=[]
for i in range(0,10):
    plt.plot(x,alpha[i,:])
    l=np.append(l,dt*i)
plt.legend(l)
plt.title("NN - vitesse variable 1")
```

On voit ici le déplacement de la masse quand on a une vitesse à 0.2. J'indique ici le temps associé à chaque itération (dt).

```python
for i in range(10,25):
    plt.plot(x,alpha[i,:])
```

J'insiste ici sur la réprésentation graphique car le passage de la masse au niveau du bord droit a été laborieux et donc ce résultat est très satisfaisant.

```python
for i in range(25,40):
    plt.plot(x,alpha[i,:])  
```

```python
for i in range(40,50):
    plt.plot(x,alpha[i,:])  
```

Avec la méthode employée on voit ici que la quantité de matière est sensiblement conservée avec une vitesse variable. Cependant on peut critiquer le lissage de nos courbes.

### Troisième cas : vitesse variable selon la quantité de matière sur la maille.

```python
f= lambda x : 1- 0.5*x
```

La vitesse varie selon la quantité de matière présente : si u = 1 on passe à une vitesse réduite de moitié. A l'inverse si u = 0 alors la vitesse passe à 1 sur la maille suivante

```python
N_t = 80
N_x = 40
dt= 1/(N_t-1)
dx=1/(N_x-1) 
x=np.linspace(0,1,N_x)
t= dt

#initialisation avec la fonction indicatrice
I_inf=0.2
I_sup=0.4
a0 = np.zeros(N_x)
a0[x>I_inf]=1 
a0[x>I_sup]=0

#Pour commencer
predict_1 =model.predict(np.append(a0[0:3],f(x[0])*dt/dx).reshape(1,4))
alpha=np.zeros((N_t,N_x))
alpha[0,:]=a0 
```

```python
n=1
cfl = max(a0)*dt/dx
print(cfl)
pred = predict_1[0,:] #les trois qté de matière pour premiere maille
stock = 0 #pred[0] Je me le permet ici car on a 0 quantité de matière dans la premiere maille.
pred = pred[1]
while cfl<1 and n<N_t-1: # La condition cfl doit être controlée pour continuer les itérations
    if n%10==1:
        print(n) #Pour suivi du temps d'exécution
    for i in range(0,N_x-2): 
        alpha[n,i+1]=alpha[n-1,i] - pred + stock      
        #Je prédis la quantité de matière sortante de la maille suivante
        stock = pred
        pred = model.predict(np.append(alpha[n-1,i:][0:3],f(alpha[n-1,i])*dt/dx).reshape(1,4))[0,:][1] #sur n-1 pour beta aussi ?
    cfl = max(alpha[n,:])*dt/dx
    #Conditions bords :Si on veut les conditions périodiques : alpha[n,0]= alpha[n,N_x-1] =0
    n=n+1
    alpha[n,0]=alpha[n-1,N_x-2] # Pour forcer le passage du bord 
    alpha[n,1]=alpha[n-1,N_x-1] 
    t+=dt
X = np.linspace(0,1,N_x)
```

```python
for i in range(0,n):
    plt.plot(X,alpha[i,:])
```

Pour ce dernier cas on observe rapidement une divergence. Le schéma et la condition prise sur la vitesse ne sont pas compatibles.
